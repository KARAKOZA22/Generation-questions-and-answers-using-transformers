```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['sentence mapping using t5 model.csv']))
df = df.drop('Unnamed: 0',axis = 1)
df
```

```python
for i in range(len(df['context'])):
  df['context'][i] = df['context'][i].replace('\n','')
  df['context'][i] = df['context'][i].replace("\\",'')
```

```python
nlp = pipeline("question-generation", model="valhalla/t5-small-qg-prepend", qg_format="prepend")
```

<h1> extracted questions and asnwers <h1>

```python
questions = []
answers = []
for i in range(len(df['context'])):
  questions.append(nlp(df['context'][i]))
questions
```

<h1> extracting questions from the dictionary <h1>

```python
t5_questions = []
t5_answers = []
w = 0
while w != 9:
  for i in questions[w]:
    for x,y in i.items():
      if x == 'question':
        s = ''.join(y)
        t5_questions.append(s)
      elif x=='answer':
        s = ''.join(y)
        t5_answers.append(s)
    w = w + 1
print(t5_questions)
print(t5_answers)
```

```python
df['question'] = t5_questions
df
```

```python
from google.colab import files

df.to_csv('QG using t5 model.csv', encoding = 'utf-8-sig') 
files.download('QG using t5 model.csv')
```

```python
df['answer'] = t5_answers
df
```
