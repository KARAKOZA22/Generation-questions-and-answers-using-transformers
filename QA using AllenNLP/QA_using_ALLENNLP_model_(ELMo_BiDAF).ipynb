```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['QG using t5 model.csv']))
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
from allennlp.predictors.predictor import Predictor
import allennlp_models.rc
predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/bidaf-elmo.2021-02-11.tar.gz")
answers = []
for (a,b) in zip(df['context'],df['question']):
  answer = (predictor.predict(
    passage=a,
    question=b))
  # function to check if score is less than 0.2 therefore, it might be incorrect
  score = max(answer["span_end_probs"])
  if score < 0.2:
    answers.append("probably not accurate answer")
  else:
    answers.append(answer['best_span_str'])
answers
```

<h1> Adding answers to DataFrame <h1>

```python
df['answers'] = answers
df
```

<h1> Downloading as a csv file <h1> 

```python
from google.colab import files
df.to_csv('QA using ALLENNLP model (ELMo-BiDAF).csv', encoding = 'utf-8-sig') 
files.download('QA using ALLENNLP model (ELMo-BiDAF).csv')
```
